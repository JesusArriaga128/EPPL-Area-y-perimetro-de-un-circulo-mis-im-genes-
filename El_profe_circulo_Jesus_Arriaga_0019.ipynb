## Formula para calcular Área

$A=\pi*r^2$

---
## Formula para calcular el perimetro
$P=2*\pi*r$

---

## Datos de entrada
$r$ ?

## Diagra de flujo en luci

```python
# Calcular Area y Perimetro de un circulo
# Variables A, P, R
# Formulas A = pi * r^ y P = 2 * pi * r
# Capturar el valor de r
import math
print("Calcular Area y Perimetro de un circulo")
r = float(input("Ingrese el radio del circulo: "))
A = math.pi * r**2
P = 2*math.pi*r
print("El area del circulo es: ", A)
print("El perimetro del circulo es: ", P)
print("Programa realizado por Jesus Arriaga NL = 5")
```
